```python
import os
import matplotlib.pyplot as plt
```

```python
precisions = ['FP16', 'FP16-INT8', 'FP32']

inf_time_fd = []
fps_fd = []
load_time_fd = []

inf_time_ld = []
fps_ld = []
load_time_ld = []

inf_time_hpe = []
fps_hpe = []
load_time_hpe = []

inf_time_ge = []
fps_ge = []
load_time_ge = []


for precision in precisions:
    precision_dir = '../metrics'    
    
    if(os.path.isdir(precision_dir) and os.path.isfile(precision_dir + '/FD/' + precision + '/stats_fd.txt')):
        with open(precision_dir + '/FD/' + precision + '/stats_fd.txt', 'r') as f:
            inf_time_fd.append(float(f.readline().split("\n")[0]))
            fps_fd.append(float(f.readline().split("\n")[0]))
            load_time_fd.append(float(f.readline().split("\n")[0]))
    else:
        inf_time_fd.append(0)
        fps_fd.append(0)
        load_time_fd.append(0)

    if(os.path.isdir(precision_dir) and os.path.isfile(precision_dir + '/' + precision + '/stats_ld.txt')):
        with open(precision_dir + '/' + precision + '/stats_ld.txt', 'r') as f:
            inf_time_ld.append(float(f.readline().split("\n")[0]))
            fps_ld.append(float(f.readline().split("\n")[0]))
            load_time_ld.append(float(f.readline().split("\n")[0]))
    else:
        inf_time_ld.append(0)
        fps_ld.append(0)
        load_time_ld.append(0)

    if(os.path.isdir(precision_dir) and os.path.isfile(precision_dir + '/' + precision + '/stats_hpe.txt')):
        with open(precision_dir + '/' + precision + '/stats_hpe.txt', 'r') as f:
            inf_time_hpe.append(float(f.readline().split("\n")[0]))
            fps_hpe.append(float(f.readline().split("\n")[0]))
            load_time_hpe.append(float(f.readline().split("\n")[0]))
    else:
        inf_time_hpe.append(0)
        fps_hpe.append(0)
        load_time_hpe.append(0)

    if(os.path.isdir(precision_dir) and os.path.isfile(precision_dir + '/' + precision + '/stats_ge.txt')):
        with open(precision_dir + '/' + precision + '/stats_ge.txt', 'r') as f:
            inf_time_ge.append(float(f.readline().split("\n")[0]))
            fps_ge.append(float(f.readline().split("\n")[0]))
            load_time_ge.append(float(f.readline().split("\n")[0]))
    else:
        inf_time_ge.append(0)
        fps_ge.append(0)
        load_time_ge.append(0)

# print(load_time_fd)
```

```python
plt.title ('Face Detection Inference')
plt.ylabel('Time (seconds)')
plt.xlabel('Model Group')
plt.bar(precisions, inf_time_fd)
```

```python
plt.title ('Face Detection Load')
plt.ylabel('Time (seconds)')
plt.xlabel('Model Group')
plt.bar(precisions, load_time_fd)
```

```python
plt.title ('Face Detection FPS')
plt.xlabel('Model Group')
plt.bar(precisions, fps_fd)
```

```python
plt.plot(precisions, inf_time_ld, 'b', marker='o', label='landmarks-regression-retail-0009')
plt.plot(precisions, inf_time_hpe, 'g', marker='o', label='head-pose-estimation-adas-0001')
plt.plot(precisions, inf_time_ge, 'r', marker='o', label='gaze-estimation-adas-0002')
plt.xlabel("Model Precision")
plt.ylabel("Time (seconds)")
plt.title('Inference Time',fontweight="bold")
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(precisions, fps_ld, 'b', marker='o', label='landmarks-regression-retail-0009')
plt.plot(precisions, fps_hpe, 'g', marker='o', label='head-pose-estimation-adas-0001')
plt.plot(precisions, fps_ge, 'r', marker='o', label='gaze-estimation-adas-0002')
plt.xlabel("Model Precision")
plt.title('Frames Per Second',fontweight="bold")
plt.legend(loc='best')
plt.show()
```

```python
plt.plot(precisions, load_time_ld, 'b', marker='o', label='landmarks-regression-retail-0009')
plt.plot(precisions, load_time_hpe, 'g', marker='o', label='head-pose-estimation-adas-0001')
plt.plot(precisions, load_time_ge, 'r', marker='o', label='gaze-estimation-adas-0002')
plt.xlabel("Precision")
plt.ylabel("Time (seconds)")
plt.title('Model Load Time',fontweight="bold")
plt.legend(loc='best')
plt.show()
```
